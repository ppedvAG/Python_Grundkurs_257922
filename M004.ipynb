### N.A.

Keine automatische Lösung -> manuell

Hier wird die string Replace Funktion verwendet

```python
data == "N.A."
```

```python
(data == "N.A.").value_counts()
```

```python
data["Fert.Rate"].replace("N.A.", "0")
```

```python
data["Fert.Rate"] = data["Fert.Rate"].replace("N.A.", "0")
```

```python
data
```

```python
def fixNA(columnName, newValue):
    data[columnName] = data[columnName].replace("N.A.", newValue)
```

```python
data.columns
```

```python
for col in data.columns:
    fixNA(col, 0)
```

```python
data
```

### Prozentzeichen

Selbes Szenario

```python
data["YearlyChange"].replace(" %", "", regex=True)  # [ ]*[%][ ]*: Regex zur Suche von beliebig vielen Abständen davor/danach
```

```python
data["YearlyChange"] = data["YearlyChange"].replace(" %", "", regex=True)
```

```python
data
```

```python
def fixPct(colName):
    data[colName] = data[colName].replace(" %", "", regex=True)
```

```python
for col in data.columns:
    fixPct(col)
```

```python
data
```

### Datentypen

Bei der Behebung von Fehlern, werden die Typen nicht automatisch angepasst

```python
data.info()
```

```python
data.convert_dtypes().info()  # Verursacht, Datentypen automatisch zu konvertieren
```

```python
data["Pop"] = data["Pop"].astype(np.int32)
```

```python
def fixType(colName, newType):
    data[colName] = data[colName].astype(newType)
```

```python
fixType("YearlyChange", np.float32)
```

```python
fixType("NetChange", np.int32)
```
