### Shortcuts

- Shift/Strg + Enter: Zelle ausführen
- Escape: Zelle verlassen (ohne Ausführung)
- A, B: Zelle oberhalb/unterhalb hinzufügen
- D + D: Zelle löschen
- X, C, V: Zelle ausschneiden, kopieren, einfügen
- M, Y: Zelle umschalten zw. Markdown und Code
- Strg + Shift + Hoch, Hinunter: Zelle verschieben
