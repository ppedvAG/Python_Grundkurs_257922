### Boolean Masken

Vektorisierung mit einer Bedingung

Ergebnis: Array mit True/False Werten

Verwendung: Filterung

```python
d = np.arange(10)
```

```python
d
```

```python
d >= 5
```

```python
bedingung = d >= 5
```

```python
d[bedingung]  # Filterung; Boolean Masken auf das originale Array anwenden
```

```python
d[d >= 5]  # Beide Schritte gemeinsam
```

```python
e = np.random.randint(100, size=(100))
```

```python
e
```

```python
# Finde alle Werte, die über dem Durchschnitt sind
avg = e.sum() / len(e)  # len auch hier möglich
avg = e.mean()
```

```python
avg
```

```python
e >= avg
```

```python
e[e >= avg]
```

```python
e[e >= avg].size
```

```python
# Finde alle überdurchschnittlichen Werte, die nicht durch 2 teilbar sind
e[(e >= avg) & ~(e % 2 == 0)]  # Hier ist and, or, not nicht möglich
```

```python
n.nbytes
```

```python
pList = list(range(10_000_000))
```

```python
pArray = np.arange(10_000_000, dtype=np.int32)
```

```python
%time sum(pList)
```

```python
%time pArray.sum()
```

```python
import time
def measureTime(func):
    start = time.time()
    func()
    end = time.time()
    print(f"{end - start}s")
```

```python
def createPythonList():
    return list(range(100_000_000))
```

```python
def createNumpyArray():
    return np.arange(100_000_000, dtype=np.int32)
```

```python
measureTime(createPythonList)
```

```python
measureTime(createNumpyArray)
```

```python
x = list(range(100_000_000))
```

```python
del x  # Daten wegwerfen -> Speicher freigeben
```

hstack, vstack

Numpy Arrays horizontal oder vertikal verknüpfen

```python
k = np.random.randint(10, size=(5, 5))
```

```python
k
```

```python
k.sum(axis=1)
```

```python
np.hstack([k, k.sum(axis=1).reshape(-1, 1)])  # WICHTIG: Benötigt immer eine Liste als Parameter
```

```python
h = np.hstack([k, k.sum(axis=1).reshape(-1, 1)])
```

```python
h.sum(axis=0)
```

```python
np.vstack([h, h.sum(axis=0)])
```
