### Globale vs. Objektorientiere Graphen

Bei globalen Graphen wird Ã¼ber globale Befehle das Diagramm konfiguriert

Um die Konfiguration abzuschlieÃŸen kann der show Befehl verwendet werden (plt.show())

```python
plt.figure(figsize=(12, 2))
plt.plot(data["Open"])
plt.xlabel("Jahr")
plt.ylabel("Preis USD")
# plt.show()  # Ohne show und figure #2, werden beide Diagramme in den selben Plot gezeichnet

# plt.figure(figsize=(12, 2))
plt.plot(data["Close"])
plt.xlabel("Jahr")
plt.ylabel("Preis USD")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)  # Hier werden die Achsen selbst angesprochen und konfiguriert

ax1.plot(data["Open"])
ax1.set_xlabel("Jahr")
ax1.set_ylabel("Preis USD")

ax2.plot(data["Open"])
ax2.set_xlabel("Jahr")
ax2.set_ylabel("Preis USD")
```

```python
fig, ((ax1_1, ax1_2), (ax2_1, ax2_2)) = plt.subplots(2, 2)
ax1_1.plot(np.random.random(100), color="red")
ax1_2.plot(np.random.random(100), color="blue")
ax2_1.plot(np.random.random(100), color="green")
ax2_2.plot(np.random.random(100), color="yellow")
```

```python
# Tortendiagramm, alle Nationen nach Top 10 sollen zusammengefasst werden
nichtTop10 = pop["WorldSharePct"][10:].sum()
top10 = np.hstack((pop["WorldSharePct"][:10], nichtTop10))

labels = np.hstack((pop.index[:10], "Rest"))

plt.pie(top10, labels=labels, autopct='%1.1f%%')

None
```
